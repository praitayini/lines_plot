```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_blsa = pd.read_excel('gms_metadata_blsa_29.xlsx', engine='openpyxl')
df_blsa = df_blsa.iloc[:29]
df_vmap = pd.read_excel('gms_metadata_vmap.xlsx',engine='openpyxl')
df_vmap = df_vmap.iloc[:29,:11]
```

```python
# sort by age and sex and assign x variable for plot
df_blsa = df_blsa.sort_values(by=['Age','Sex'])
df_blsa = df_blsa.assign(x='BLSA')

df_vmap = df_vmap.sort_values(by=['Age', 'Sex'])
df_vmap = df_vmap.assign(x='VMAP')
```

```python
df_cat =  pd.concat([df_vmap,df_blsa])
```

```python
df_cat.head(1)
```

```python
plt.rcParams.update({'font.size':37})
fig, axes = plt.subplots(1,4, figsize=(35,15))#, gridspec_kw = {'width_ratios': [1,0.4,1,0.4,1,0.4]})

flierprops = dict(markerfacecolor='0.75', markersize=5, marker='.',
              linestyle='none')
palette= {'VMAP': 'indianred', 'BLSA': 'steelblue'}

box_plot = sns.swarmplot(ax = axes[0], x="x",y='Modularity',data=df_cat, palette=palette, size=20)
axes[0].set_xlabel(' ')

# to draw the lines
idx0 = 0
idx1 = 1
locs1 = axes[0].get_children()[idx0].get_offsets()
locs2 = axes[0].get_children()[idx1].get_offsets()
sort_idxs1 = np.argsort(df_vmap['Modularity'].tolist())
sort_idxs2 = np.argsort(df_blsa['Modularity'].tolist())
locs2_sorted = locs2[sort_idxs2.argsort()][sort_idxs1]
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2_sorted[i, 0]]
    y = [locs1[i, 1], locs2_sorted[i, 1]]
    axes[0].plot(x, y, color="black", alpha=0.5)
    
    
box_plot = sns.swarmplot(ax = axes[1], x="x",y='Global Efficiency',data=df_cat, palette=palette, size=20)
axes[1].set_xlabel(' ')
idx0 = 0
idx1 = 1
locs1 = axes[1].get_children()[idx0].get_offsets()
locs2 = axes[1].get_children()[idx1].get_offsets()
sort_idxs1 = np.argsort(df_vmap['Global Efficiency'].tolist())
sort_idxs2 = np.argsort(df_blsa['Global Efficiency'].tolist())
locs2_sorted = locs2[sort_idxs2.argsort()][sort_idxs1]
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2_sorted[i, 0]]
    y = [locs1[i, 1], locs2_sorted[i, 1]]
    axes[1].plot(x, y, color="black", alpha=0.5)
    
box_plot = sns.swarmplot(ax = axes[2], x="x",y='Avg Bet Centrality',data=df_cat, palette=palette, size=20)
axes[2].set_xlabel(' ')
idx0 = 0
idx1 = 1
locs1 = axes[2].get_children()[idx0].get_offsets()
locs2 = axes[2].get_children()[idx1].get_offsets()
sort_idxs1 = np.argsort(df_vmap['Avg Bet Centrality'].tolist())
sort_idxs2 = np.argsort(df_blsa['Avg Bet Centrality'].tolist())
locs2_sorted = locs2[sort_idxs2.argsort()][sort_idxs1]
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2_sorted[i, 0]]
    y = [locs1[i, 1], locs2_sorted[i, 1]]
    axes[2].plot(x, y, color="black", alpha=0.5)
    
box_plot = sns.swarmplot(ax = axes[3], x="x",y='Char Path Length',data=df_cat, palette=palette, size=20)
axes[3].set_xlabel(' ')
idx0 = 0
idx1 = 1
locs1 = axes[3].get_children()[idx0].get_offsets()
locs2 = axes[3].get_children()[idx1].get_offsets()
sort_idxs1 = np.argsort(df_vmap['Char Path Length'].tolist())
sort_idxs2 = np.argsort(df_blsa['Char Path Length'].tolist())
locs2_sorted = locs2[sort_idxs2.argsort()][sort_idxs1]
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2_sorted[i, 0]]
    y = [locs1[i, 1], locs2_sorted[i, 1]]
    axes[3].plot(x, y, color="black", alpha=0.5)
plt.subplots_adjust(wspace=0.6)#,left=0.05,right=0.99)
```
